```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Plotting all images
plt.figure(figsize=(15, 15))

# 1) Plotting the original image
# Read the original image
img = cv2.imread("3.jpg")
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(1, 3, 1)
plt.imshow(new_img)
plt.title('Original Image')
plt.axis('off')

# 2) Define a kernel for filtering
k1 = np.array([[-1, -1, -1],[-1,  9, -1],[-1, -1, -1]])
# Apply the filter to the original image
filtered_img = cv2.filter2D(img, -1, k1)
# Plotting the filtered image
plt.subplot(1, 3, 2)
plt.imshow(filtered_img)
plt.title('Filtered Image')
plt.axis('off')

#3) blur image 
blurred_img = cv2.blur(new_img, (5,5))
# Plotting the blurred image
plt.subplot(1, 3, 3)
plt.imshow(blurred_img)
plt.title('Blurred Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
