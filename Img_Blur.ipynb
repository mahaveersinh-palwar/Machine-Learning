```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("1.jpg")

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

blur = cv2.GaussianBlur(img,(5,5),0)

blur2 = cv2.medianBlur(img,3)

plt.subplot(1,4,1),plt.imshow(img),plt.title("First")

plt.subplot(1,4,2),plt.imshow(blur),plt.title("First")

plt.subplot(1,4,3),plt.imshow(blur2),plt.title("First")

plt.subplot(1,4,4),plt.imshow(blur2),plt.title("First")
```
